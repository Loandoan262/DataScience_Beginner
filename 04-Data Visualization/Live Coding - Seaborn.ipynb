## Grids

```python
iris = sns.load_dataset("iris")
```

```python
iris
```

```python
grid = sns.PairGrid(iris)
grid.map(plt.scatter)
```

```python
grid = sns.PairGrid(iris)
grid.map_diag(sns.histplot)
grid.map_lower(sns.kdeplot)
grid.map_upper(sns.scatterplot)
```

## Facet Grid

```python
tips
```

```python
sns.FacetGrid(data = tips, col = "time", row = "smoker")
```

```python
f = sns.FacetGrid(data = tips, col = "time", row = "smoker")
f.map(plt.hist, "total_bill")
```

```python
f = sns.FacetGrid(data = tips, col = "time", row = "smoker", hue = "sex")
f.map(plt.scatter, "total_bill", "tip", marker = "v").add_legend()
```

## JointGrid

```python
j = sns.JointGrid(x="total_bill", y = "tip", data = tips)
j.plot(sns.regplot, sns.histplot)
```

```python
sns.jointplot(x="total_bill", y = "tip", data=tips, kind = "reg")
```

## Regression Plots

```python
tips
```

```python
sns.lmplot(x= "total_bill", 
           y = "tip", 
           data = tips, 
           hue = "sex", 
           legend=True, 
           palette = "Set1", 
           markers = ["o", "v"],
           scatter_kws={"s": 10}
          )
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = tips, col="sex")
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = tips, row="sex", col = "time")
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = tips, col = "day", hue = "sex", aspect = .5, height = 10)
```
