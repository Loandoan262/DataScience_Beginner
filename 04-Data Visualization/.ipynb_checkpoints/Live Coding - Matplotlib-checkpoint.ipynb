## Matplotlib Object Oriented Method 

```python
alpha = np.linspace(0,5,11)
beta = alpha**5
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(alpha, beta, "g")
axes.plot(alpha, alpha**3, "b")
axes.set_xlabel("Set x label")
axes.set_ylabel("Set y label")
axes.set_title("Set Title")
```

```python
# Giong cach viet cua phan Basic
#plt.plot(alpha, beta, "g")
#plt.plot(alpha, alpha**3, "b")
plt.plot(alpha, beta, "g", alpha, alpha**3, "b")
```

```python
## Example 2
fig_new = plt.figure()

axes1 = fig_new.add_axes([0.1,0.1,.8,.8])
axes2 = fig_new.add_axes([.8, .8, .4, .3])

axes1.plot(alpha, beta, "y--")
axes1.set_xlabel("Alpha label 1")
axes1.set_ylabel("Beta label 1")

axes2.plot(beta,alpha, "g.")
axes1.set_xlabel("Beta label 2")
axes1.set_ylabel("Alpha label 2")
```

## Legends, labels, titles

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(alpha, alpha**2, label="$\u03B1^2$".format())
ax.plot(alpha, alpha**3, label=f"$alpha^3$")
ax.legend()
```

```python
pic, axes = plt.subplots()
#RGB hex code
axes.plot(alpha, beta, c = "#bebebe", marker="o", lw=3, ls="--")
```

## Plot range

```python
pic, axes = plt.subplots(1, 3, figsize = (12,4))

axes[0].plot(alpha, beta, alpha, alpha**3)
axes[0].set_title("default")

axes[1].plot(alpha, beta, alpha, alpha**3)
axes[1].axis("tight")
axes[1].set_title("tight")

axes[2].plot(alpha, beta, alpha, alpha**3)
axes[2].set_ylim([200, 4000])
axes[2].set_xlim([2, 4.5])
```
