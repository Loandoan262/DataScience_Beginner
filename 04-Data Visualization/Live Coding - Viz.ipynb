## Matplotlib Object Oriented Method 

```python
alpha = np.linspace(0,5,11)
beta = alpha**5
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(alpha, beta, "g")
axes.plot(alpha, alpha**3, "b")
axes.set_xlabel("Set x label")
axes.set_ylabel("Set y label")
axes.set_title("Set Title")
```

```python
#plt.plot(alpha, beta, "g")
#plt.plot(alpha, alpha**3, "b")
plt.plot(alpha, beta, "g", alpha, alpha**3, "b")
```

```python
## Example 2
fig_new = plt.figure()

axes1 = fig_new.add_axes([0.1,0.1,.8,.8])
axes2 = fig_new.add_axes([.8, .8, .4, .3])

axes1.plot(alpha, beta, "y--")
axes1.set_xlabel("Alpha label 1")
axes1.set_ylabel("Beta label 1")

axes2.plot(beta,alpha, "g.")
axes1.set_xlabel("Beta label 2")
axes1.set_ylabel("Alpha label 2")
```
